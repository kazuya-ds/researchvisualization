Tables for Papers and Posters

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# --- 1. CONFIGURATION: Color Palette (Electric Blue & Grey) ---
PRIMARY_COLOR = '#1A73E8'   # Modern AI Blue
ACCENT_COLOR = '#E8F0FE'    # Soft background accent
TEXT_COLOR = '#202124'      # Near-black for readability
HEADER_BG = '#F1F3F4'       # Subtle grey for headers

# --- 2. DATA PREPARATION ---
data = {
    "Category": [
        "Resource Efficiency",
        "Feature Precision",
        "Trust & XAI",
        "Sustainability"
    ],
    "Industrial Reasoning": [
        "Optimal Trade-off: Lightweight architecture (B0) designed for edge-deployment\nwithout sacrificing accuracy (9497%).",
        "Compound Scaling: Balanced depth and resolution capture micro-level textural\nanomalies missed by shallower networks.",
        "Decision Support: SHAP/Grad-CAM integration transforms black-box predictions\ninto actionable maintenance interventions.",
        "Operational Scaling: Energy-efficient inference supports long-term net-zero goals\nin continuous monitoring systems."
    ]
}

df = pd.DataFrame(data)

# --- 3. GRAPH DESIGN: Infographic Table Layout ---
fig, ax = plt.subplots(figsize=(14, 6))
ax.axis('off')

# Design concepts: Central alignment, bold headers, and high contrast
table = ax.table(
    cellText=df.values,
    colLabels=[f"$\mathbf{{{c}}}$" for c in df.columns], # Bold LaTeX headers
    loc='center',
    cellLoc='left'
)

# --- 4. VISUAL STYLING: Refined Poster Aesthetics ---
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1, 4.5)  # Increased height for "Visual Rest" (White Space)

for (row, col), cell in table.get_celld().items():
    cell.set_edgecolor('#DDDDDD')  # Soft borders
    cell.set_linewidth(0.5)

    if row == 0: # Header styling
        cell.set_text_props(weight='bold', color='white', fontsize=13)
        cell.set_facecolor(PRIMARY_COLOR)
        cell.set_edgecolor(PRIMARY_COLOR)
    elif col == 0: # Category Column Styling (Hierarchy)
        cell.set_text_props(weight='bold', color=PRIMARY_COLOR)
        cell.set_facecolor(ACCENT_COLOR)
    else: # Description Styling
        cell.set_facecolor('white')
        cell.set_text_props(color=TEXT_COLOR)

plt.title('Why EfficientNet-B0? Key PdM Decision Drivers',
          fontsize=18, fontweight='bold', color=PRIMARY_COLOR, pad=30)

# --- 5. EXPORT & DOWNLOAD (High-Res 600 DPI for Print) ---
output_file = 'Poster_Table_.png'
plt.savefig(output_file, dpi=600, bbox_inches='tight', facecolor='white')
plt.show()

files.download(output_file)
print(f"Poster Table saved as: {output_file}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from google.colab import files # For automatic download in 2026

def create_poster_infographic():
    # 1. SETUP: Professional "Micro-Industrial" Palette
    TECH_BLUE = '#1A73E8'   # Primary AI brand color
    STEEL_GREY = '#3C4043'  # Neutral industrial base
    SUCCESS_GREEN = '#34A853'
    WARNING_YELLOW = '#FBBC04'
    DANGER_RED = '#EA4335'
    BG_SOFT = '#F8F9FA'     # Trending "off-white" background

    fig, ax = plt.subplots(figsize=(16, 9), facecolor='white')
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.axis('off')

    # HEADER: High-impact Title Bar
    ax.add_patch(patches.Rectangle((0, 85), 100, 15, color=TECH_BLUE, alpha=0.1, zorder=0))
    plt.text(50, 93, "EFFICIENTNET-B0: TRANSFER LEARNING FOR INDUSTRIAL PdM",
             ha='center', fontsize=26, fontweight='bold', color=TECH_BLUE)
    plt.text(50, 89, "Optimizing Steel Wear Detection for Manufacturing IoT Hardware",
             ha='center', fontsize=14, style='italic', color=STEEL_GREY)

    # --- SECTION 1: TRANSFER LEARNING SOURCE (THE "INPUT") ---
    ax.add_patch(patches.FancyBboxPatch((5, 50), 20, 25, boxstyle="round,pad=0.3", fc=BG_SOFT, ec=TECH_BLUE))
    plt.text(15, 68, "ImageNet\n(Pre-trained)", ha='center', va='center', fontsize=12, weight='bold')
    plt.text(15, 58, "General Feature\nExtraction (Edges/Textures)", ha='center', va='center', fontsize=9)

    # Visual Connector: Transfer Arrow
    ax.annotate('', xy=(32, 62.5), xytext=(26, 62.5), arrowprops=dict(arrowstyle='->', lw=3, color=TECH_BLUE))

    # --- SECTION 2: EFFICIENTNET-B0 CORE (THE "ENGINE") ---
    plt.text(50, 78, "IOT-OPTIMIZED CLASSIFIER", fontweight='bold', fontsize=12, color=STEEL_GREY, ha='center')
    ax.add_patch(patches.FancyBboxPatch((35, 45), 30, 35, boxstyle="round,pad=0.5", fc=TECH_BLUE, alpha=0.9))
    plt.text(50, 68, "B0 Architecture", ha='center', color='white', fontsize=18, weight='bold')
    plt.text(50, 58, "Compound Scaling:\nDepth  Width  Resolution", ha='center', color='white', fontsize=10)
    plt.text(50, 50, "OPTIMIZED FOR\nManufacturing Edge Devices", ha='center', color='white', fontsize=10, weight='bold')


    # Visual Connector: Output Arrow
    ax.annotate('', xy=(74, 62.5), xytext=(66, 62.5), arrowprops=dict(arrowstyle='->', lw=3, color=TECH_BLUE))

    # --- SECTION 3: WEAR STATES (THE "DIAGNOSTICS") ---
    plt.text(85, 78, "4-STATE DIAGNOSTICS", fontweight='bold', fontsize=12, color=STEEL_GREY, ha='center')

    states = [
        ("PERFECT", SUCCESS_GREEN, "No Defects"),
        ("LIGHT", WARNING_YELLOW, "Surface Abrasion"),
        ("MODERATE", '#FF9800', "Visible Dents"),
        ("HEAVY", DANGER_RED, "Structural Risk")
    ]

    for i, (name, color, detail) in enumerate(states):
        y_pos = 70 - (i * 14)
        ax.add_patch(patches.Circle((82, y_pos), 4, color=color, alpha=0.8))
        plt.text(82, y_pos, str(i+1), ha='center', va='center', color='white', weight='bold')
        plt.text(88, y_pos+1, name, va='center', fontsize=11, weight='bold', color=STEEL_GREY)
        plt.text(88, y_pos-2, detail, va='center', fontsize=8, color=STEEL_GREY, alpha=0.7)

    # # --- FOOTER: Performance Bar ---
    # ax.add_patch(patches.Rectangle((0, 0), 100, 15, color=STEEL_GREY, alpha=0.05))
    # stats = [("ACCURACY", "96.4%"), ("LATENCY", "12ms"), ("FLOPs", "0.39B"), ("IoT NODES", "Ready")]
    # for i, (label, val) in enumerate(stats):
    #     x_pos = 20 + (i * 20)
    #     plt.text(x_pos, 8, label, ha='center', fontsize=9, color=STEEL_GREY, weight='bold')
    #     plt.text(x_pos, 3, val, ha='center', fontsize=16, weight='bold', color=TECH_BLUE)

    plt.tight_layout()
    save_path = 'PdM_EfficientNetB0_Visualization.png'
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()

    # Automatically download for local use
    files.download(save_path)

# Run visualization
create_poster_infographic()
```
